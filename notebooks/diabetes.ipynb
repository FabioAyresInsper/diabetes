```python
import pathlib

filename = pathlib.Path.cwd().parent / 'data' / 'visceral_fat.csv'
filename
```

```python
import pandas as pd

data = pd.read_csv(filename)
```

```python
data
```

```python
def converte_idade_gestacional(texto):
    partes_do_texto = texto.split(',')
    semanas = int(partes_do_texto[0])
    dias = int(partes_do_texto[1])
    return 7*semanas + dias

data['current gestational age'] = data['current gestational age'].apply(converte_idade_gestacional)
data['gestational age at birth'] = data['gestational age at birth'].apply(converte_idade_gestacional)
```

```python
data
```

```python
data.info()
```

```python
data = data.drop(columns=['number', 'diabetes mellitus'])
```

```python
data.info()
```

```python
data = data.dropna(axis=0)
```

```python
data
```

```python
X = data.drop(columns=['gestational dm']).copy()
y = data['gestational dm'].copy()
```

```python
import seaborn as sns

corr = X.corr()

sns.heatmap(corr)
```

```python
X.corrwith(y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train_valid, X_test_valid, y_train_valid, y_test_valid = train_test_split(X_train, y_train, test_size=0.33, random_state=42)
```

Modelo 1: Logistic regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model = LogisticRegression(solver='liblinear', C=1e-6)
model.fit(X_train_valid, y_train_valid)
y_pred = model.predict(X_test_valid)

accuracy = accuracy_score(y_test_valid, y_pred)
accuracy
```

```python
y.value_counts(normalize=True)
```

Modelo 2: decision tree

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X_train_valid, y_train_valid)
y_pred = model.predict(X_test_valid)

accuracy = accuracy_score(y_test_valid, y_pred)
accuracy
```

Modelo 3: Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train_valid, y_train_valid)
y_pred = model.predict(X_test_valid)

accuracy = accuracy_score(y_test_valid, y_pred)
accuracy
```
